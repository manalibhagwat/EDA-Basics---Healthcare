## Data Cleaning - 

```python
# Change the data types for dates - 

df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.columns[1:]
```

```python
agg_func = {}

for col in df.columns[1:]:
    if np.issubdtype(df[col].dtype, np.number):
        agg_func[col] = 'sum'
    else:
        agg_func[col] = 'first'
```

```python
agg_func
```

```python
# Converting Age to first instead of sum

agg_func['Age'] = 'first'
```

```python
agg_func
```

```python
df.Name.value_counts()
```

```python
df1 = df.groupby('Name').aggregate(agg_func)
df1 = df1.reset_index()
df1.shape
```

```python
# Sanity check - 

df1.Name.value_counts()
```

```python
df1.describe()
```

```python
# Calculating days in hospital - 

df1['Days in Hosp'] = (df1['Discharge Date'] - df1['Date of Admission']).dt.days
```

```python
# Extracting year and month from date columns - 

df1['Admission Year'] = pd.DatetimeIndex(df1['Date of Admission']).year
df1['Admission Month'] = pd.DatetimeIndex(df1['Date of Admission']).month

df1['Discharge Year'] = pd.DatetimeIndex(df1['Discharge Date']).year
df1['Discharge Month'] = pd.DatetimeIndex(df1['Discharge Date']).month
```

```python
df1.head()
```

```python
# Per day rate based on billing amount and days spent in hospital pr patient

df1['Rate per Day'] = df1['Billing Amount']/df1['Days in Hosp']
```

```python
# Creating a new dataframe by dropping unnecessary rows - 

df2 = df1.drop(['Name','Date of Admission', 'Discharge Date', 'Room Number'], axis=1)
df2.head()
```

```python
df2.groupby('Admission Month')['Billing Amount'].mean().plot()
plt.show()
```

- The graph shows a sudden decline in the billing values in various months and a steep incline as well

## Bivariate Analysis

```python
# Analysing relationship between Billing Amount & Number of days in hospital

sns.scatterplot(data=df1, x='Billing Amount', y='Days in Hosp')
plt.show()
```

- There is no clear pattern, although logic dictates a positive correlation and r value

```python
# Analysing categorical variables across Ages

res = pd.pivot_table(data=df2, values='Age', columns='Admission Type', index='Medical Condition', 
                     aggfunc='mean')

res.plot(kind='bar', stacked=True, figsize=[10,6])
plt.show()
```

- The age wise analysis has not yielded any clear results as all medical conditions across admission types seem to have occurred in similar age ranges

```python
# Analysing categorical variables for Billing amounts

res = pd.pivot_table(data=df2, values='Billing Amount', columns='Admission Type', index='Medical Condition', 
                     aggfunc='mean')

res.plot(kind='bar', stacked=True, figsize=[10,6])
plt.show()
```

- Asthama, diabetes, and obesity have a higher billing amount due to urgent admission type as compared to other medical conditions

```python
# Analysing categorical variables across billing amount - heatmap

res = pd.pivot_table(data=df2, values='Billing Amount', columns='Test Results', index='Medical Condition', 
                     aggfunc='mean')

sns.heatmap(res, annot=True, cmap='RdYlGn')
plt.show()
```

```python
# Analysing numeric vales correlating with each other - 

sns.pairplot(data=df2, vars=['Billing Amount', 'Age', 'Days in Hosp'])
plt.show()
```

- No clear patterns are derivable from these graphs
